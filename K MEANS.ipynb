```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler 
import matplotlib.pyplot as plt
```

```python
#1. Load data from excel
file_path="D:/CODING/PYTHON/DWDM ALGO/sample_kmeans_dataset.xlsx" 
df = pd.read_excel (file_path)
```

```python
#2. select only numeric values
num_df = df.select_dtypes (include=['float64', 'int64'])
```

```python
#3. Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(num_df)
```

```python
#4. K-means clustering
k = 3
kmeans = KMeans (n_clusters = k, random_state = 1)
kmeans.fit(scaled_data)
```

```python
#5. add cluster Labels to dataframe 
df['Cluster'] = kmeans.labels_
```

```python
#6. Plot the clusters
if scaled_data.shape[1] >= 2: 
    plt.figure(figsize=(8,6)) 
plt.scatter(scaled_data[:,0], scaled_data[:,1], c = kmeans.labels_, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],c='red', s=200, alpha=0.75, marker = 'X', label='Centroid')
plt.title("K-Means Clustering (Feat.1 vs Feat.2)") 
plt.xlabel("Feature 1 (scaled)") 
plt.ylabel("Feature 2 (scaled)")
plt.grid(True) 
plt.show()
```
