```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
# sample data 
# each transactions is the list of items brought togther 
# for example ["I1","I2","I3"] means a transaction cointaning items I1, I2 and I3
data = [
    ["I1","I2","I5"],
    ["I2","I4"],
    ["I2","I3"],
    ["I1","I2","I4"],
    ["I1", "I3"],
    ["I1", "I3"],
    ["I1", "I3"],
    ["I1","I2","I3", "I5"],
    ["I1","I2","I3"]
]
```

```python
type(data)
```

```python
te = TransactionEncoder()
transformed_data = te.fit(data).transform(data)
```

```python
print(transformed_data)
```

```python
df = pd.DataFrame(transformed_data, columns = te.columns_)
```

```python
print(df)
```

```python
# get user input for minimmum support as absolute count
min_support_count = int(input("ENTER MINIMMUM SUPPORT COUNT(ABSOLUTE NUMBER OF TRANSACTIONS, e.g.. 3 FOR 3 TRANSACTIONS):"))

# convert absolute support count to relative support
min_support = min_support_count / len(data)
```

```python
frequent_itemsets = apriori(df,min_support = min_support, use_colnames = True)
print("\nAll Frequent Itemsets:")
print(frequent_itemsets)
```

```python
# Get user input for minimmum confidence
min_confidence = float(input("ENTERR MINIMMUM CONFIDENCE (AS A DECIMAL e.g., 0.6 FOR 60%: "))
```

```python
# Example of association_rules function usage:
# association_rules(frequent_itemsedts, metric:"confidence", min_threshold = 0.6)
# - frequent_itemsets: DataFrame containing frequent itemsets and their support values
# - metric: Evaluation metric for the rules(e.g., "confidence")
# - min_threshold: Minimmum threshold for the choosen metric
# Returns a DataFrame containing the association rules with their support, confidence, and lift

# Generating all association rules from the frequent datasets
# Filters rules bsaed on confidence threshold
rules = association_rules(frequent_itemsets, metric = "confidence" , min_threshold = min_confidence)

# Display all rules, not just the maximmum itemset rules
print("\nAll association Rules:")
print(rules[['antecedents','consequents','support','confidence']])
```
